```
class Array2D: #Clase Array2D
	def __init__(self,ren,col):
		self.reng = ren
		self.colu = col
		self.array=[ [0 for i in range(col)] for i in range(ren)]

	def clear(self,dato):
		self.array =[ [[dato]for i in range(self.colu)] for i in range(self.reng)]

	def get_row_size(self):
		return self.reng

	def get_col_size(self):
		return self.colu

	def set_item(self,ren,col,dato):
		self.array[ren][col] = dato

	def get_item(self,ren,col):
		return self.array[ren][col]

	def to_string(self):
		return self.array


class JuegoDeLaVida:
	def __init__(self,rens,cols,generacion,poblacion):
		self.largo = cols
		self.alto = rens
		self.grid = Array2D(self.alto,self.largo)
		self.generaciones = generacion

		for cel in poblacion:
			self.grid.set_item(cel[0],cel[1],1)		

	def imprime_grid(self):
		for r in range(self.grid.get_col_size()):
			for c in range(self.grid.get_row_size()):
				if self.grid.get_item(r,c) == 0:

					print('░░',end=' ')
				else:
					print('██',end =' ')

			print('')
		print('')

	def get_num_rows(self):
		return self.alto

	def get_num_cols(self):	
		return self.largo

	def set_celula_muerta(self,ren,col):
		self.grid.set_item(ren,col,0)
  	
	def set_celula_viva(self,ren,col):
		self.grid.set_item(ren,col,1)	

	def get_is_viva(self,ren,col):
		return self.grid.get_item(ren,col) == 1

	def get_is_muerta(self,ren,col):
		return self.grid.get_item(ren,col) == 0

	def get_numero_vecinos_vivos(self,rens,cols): 
		contador = 0

		for uwu in [rens-1, rens, rens+1]:
			for owo in [cols-1, cols, cols+1]:
				try:
					if self.grid.get_item(uwu,owo) == 1  and (uwu,owo) != (rens,cols):
						contador += 1
				except Exception as e:
					pass
		return contador

	def evolucionar(self): 
		for i in range(self.generaciones):
			self.imprime_grid()

			l_v = []
			l_m = []
			for renglon in range(self.alto):
				for columna in range(self.largo):
					if self.get_numero_vecinos_vivos(renglon,columna) == 3:
						l_v.append((renglon,columna))

					if self.get_numero_vecinos_vivos(renglon,columna) < 2 or self.get_numero_vecinos_vivos(renglon,columna) > 3:
						l_m.append((renglon,columna))
      
			for x in l_v:
				self.set_celula_viva(x[0],x[1])
			for y in l_m:
				self.set_celula_muerta(y[0],y[1])





prueba = JuegoDeLaVida(6,6,7,[(1,2),(2,1),(2,2),(2,3)])
prueba.evolucionar()
```
