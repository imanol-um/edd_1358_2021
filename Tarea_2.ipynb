```
"""
Array
"""
class Array:
    def __init__(self,tam):
        self.__datos = [0 for x in range(tam)]
    def get_item(self, index):
        dato = 0
        if index >= 0 and index < len(self.__datos):
            dato = self.__datos[index]
        else:
            print("Error en el indice")
        return dato

    def set_item(self, dato, index):
        if index >= 0 and index < len(self.__datos):
            self.__datos[index] =   dato 
        else:
            print("Error en el indice")
    
    def get_Length(self):
        return len(self.__datos)

    def clear(self, dato):
        self.datos = [dato for x in range(self.__tam)]

    def _getIterator(self):
        return _getIterator(self.__datos)

class _getIterator:
    def __init__(self,arreglo):
        self.__arreglo = arreglo
        self.__indice_actual = 0
    
    def __iter__ ( self ):
        return self
    def __next__ ( self ):
        if self.__indice_actual < len(self.__arreglo):
            dato = self.__arreglo[self.__indice_actual]
            self.__indice_actual += 1
            return dato
        else:
            raise StopIteration
```

```
class Empleado:
    def __init__( self, num, nom, pat, mat, he, sb, anio):
        self.__num_trabajador = num
        self.__nombres = nom
        self.__paterno = pat
        self.__materno = mat
        self.__horas_extra = he
        self.__sueldo_base = sb
        self.__anio_ingreso = anio  

    @property
    def num_trabajador(self):
        return self.__num_trabajador
    @num_trabajador.setter
    def num_trabajador(self, num):
        self.__num_trabajador = float(num)

    @property
    def nombres(self):
        return self.__nombres
    @nombres.setter
    def nombres(self, nom):
        self.__nombres = nom

    @property
    def paterno(self):
        return self.__paterno
    @paterno.setter
    def paterno(self, pat):
        self.__paterno = pat
    
    @property
    def materno(self):
        return self.__materno
    @materno.setter
    def materno(self, mat):
        self.__materno = mat
    
    @property
    def horas_extra(self):
        return self.__horas_extra
    @horas_extra.setter
    def horas_extra(self, he):
        self.__horas_extra = float(he)
    
    @property
    def sueldo_base(self):
        return self.__sueldo_base
    @sueldo_base.setter
    def sueldo_base(self, sb):
        self.__sueldo_base = float(sb)

    @property
    def anio_ingreso(self):
        return self.__anio_ingreso
    @anio_ingreso.setter
    def anio_ingreso(self, anio):
        self.__anio_ingreso = float(anio)
     
    def to_String ( self, x): 
        if x == "numero de trabajador":
            return "numero de trabajador = " + self.__num_trabajador  
        
        elif x == "nombres":
            return "nombres = " + self.__nombres
        
        elif x == "paterno":
            return "paterno = " + self.__paterno
        
        elif x == "materno":
            return "materno = " + self.__materno

        elif x == "horas extra":
            return "horas extra = " + self.__horas_extra

        elif x == "sueldo_base":
            return "sueldo base = " + self.__sueldo_base
        
        elif x == "año ingreso":
            return "año ingreso = " + self.__anio_ingreso
        else:
            print("Asignacion no valida")  
        

```
