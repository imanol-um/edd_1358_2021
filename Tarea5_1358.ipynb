```
class NodoDoble:
    def __init__( self , value , previo = None , siguiente = None ):
        self.data = value
        self.prev = previo
        self.next = siguiente

class DoubleLinkedList:
    def __init__( self ):
        self.__head = None
        self.__tail = None
        self.__size = 0

    def get_size( self ):
        return self.__size

    def is_empty( self ):
        return self.__size == 0

    def append( self, value ):
        if self.is_empty():
            nuevo_nodo = NodoDoble(value)
            self.__head = nuevo_nodo
            self.__tail = nuevo_nodo
        else:
            nuevo_nodo = NodoDoble(value, self.__tail, None)
            self.__tail.next = nuevo_nodo
            self.__tail = nuevo_nodo
        self.__size += 1

    def find_from_tail( self, value ):
        curr_node = self.__tail
        contador = self.__size
        while curr_node != None:
            if curr_node.data == value:
                return print(f"El valor {value} se encuentra en la posición: {contador}")
            if curr_node.prev == None:
                return print("El valor no se encuentra en la lista")
            curr_node = curr_node.prev
            contador -= 1

    def find_from_head( self, value ):
        curr_node = self.__head
        contador = 0
        while curr_node != None:
            contador += 1
            if curr_node.data == value:
                return print(f"El valor {value} se encuentra en la posición: {contador}")
            if curr_node.next == None:
                return print("El valor no se encuentra en la lista")
            curr_node = curr_node.next

    def remove_form_tail( self, value ):
        if self.is_empty():
            print("La lista esta vacia.")
            pass
        elif value == self.__tail.data:
            curr_node = self.__tail
            self.__tail = curr_node.prev 
            curr_node.prev.next = None
        else:
            curr_node = self.__tail
            while curr_node.data != value and curr_node != None:
                curr_node = curr_node.prev
                if curr_node.data == value:
                    curr_node.prev.next = curr_node.next
                    curr_node.next.prev = curr_node.prev
        self.__size -= 1

    def remove_form_head( self, value ):
        if self.is_empty():
            print("La lista esta vacia.")
            pass
        elif value == self.__head.data:
            curr_node = self.__head
            self.__head = curr_node.next 
            curr_node.next.prev = None
        else:
            curr_node = self.__head
            while curr_node.data != value and curr_node != None:
                curr_node = curr_node.next
                if curr_node.data == value:
                    curr_node.prev.next = curr_node.next
                    curr_node.next.prev = curr_node.prev
        self.__size -= 1

    def transversal ( self ):
        curr_node = self.__head
        while curr_node != None:
            print(f"<--- {curr_node.data} ----->", end="")
            curr_node = curr_node.next
        print("")

    def reverse_transversal( self ):
        curr_node = self.__tail
        while curr_node != None:
            print(f"<--- {curr_node.data} ----->", end="")
            curr_node = curr_node.prev
        print("")
```

```
ldl = DoubleLinkedList()

print(ldl.get_size())

ldl.append(1)
print(ldl.get_size())
ldl.transversal()
ldl.append(2)
ldl.append(3)
ldl.transversal()
ldl.remove_form_head(2)
ldl.transversal()
ldl.remove_form_head(1)
ldl.transversal()
ldl.append(2)
ldl.append(1)
ldl.transversal()
ldl.reverse_transversal()
ldl.find_from_head(3)
ldl.find_from_head(1)
ldl.find_from_tail(3)
ldl.find_from_tail(1)
```
